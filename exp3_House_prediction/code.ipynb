```python
import matplotlib.pyplot as plt 

import seaborn as sns 

import plotly.express as px
```

```python
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
plt.figure(figsize=(8,5))  # Set figure size
plt.plot(df["area"],df["prices"],marker='o',linestyle="-", color="blue")
plt.xlabel("area")   # X-axis label= means the trend is plotted over the order of rows in the dataset.
plt.ylabel("price")  # Y-axis label
plt.title("House_price vs area")  # Title of the plot
plt.show()  # Display the plot
```

```python
plt.scatter(df["area"],df["prices"],marker="+",color="red")
plt.xlabel("area")   # X-axis label= means the trend is plotted over the order of rows in the dataset.
plt.ylabel("price")  # Y-axis label
plt.title("House_price vs area")  # Title of the plot
plt.show()  # Display the plot
```
