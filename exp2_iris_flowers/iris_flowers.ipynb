### Exploratory Analysis and Visualisation

```python
import matplotlib.pyplot as plt #basic drawing tool.helps create charts, graphs, and plots like line graphs, bar charts, and histograms.

import seaborn as sns #It builds on Matplotlib and makes prettier, more informative graphs.great for visualizing complex relationships in data with heatmaps, box plots, etc.

import plotly.express as px #It makes interactive graphs, so you can zoom in, hover over points, and explore data dynamically.
```

```python
print(iris_df.head())
```

```python
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# line plot = to visualize trends over an index
plt.figure(figsize=(8,5))  # Set figure size
plt.plot(iris_df.index, iris_df["sepal_length"],marker='o',linestyle="-", color="blue", label="Sepal Length")
plt.xlabel("Index")   # X-axis label= means the trend is plotted over the order of rows in the dataset.
plt.ylabel("Sepal Length (cm)")  # Y-axis label
plt.title("Line Plot of Sepal Length")  # Title of the plot
plt.legend()  # Show legend
plt.show()  # Display the plot
```

```python
# bar plot = used to compare categorical data.
sns.barplot(x="species", y="sepal_length", data=iris_df)  
plt.title("Bar Plot of Sepal Length by Species")  
plt.xlabel("Species")  
plt.ylabel("Sepal Length (cm)")  
plt.show()
```

```python
# Histogram = shows the distribution (spread) of data.
plt.hist(iris_df["sepal_length"], bins=10, color="skyblue", edgecolor="black") #bins=10 divides the data into 10 intervals, 
# color="skyblue" sets the color of bars, edgecolor="black" outlines each bar.
plt.title("Histogram of Sepal Length")  
plt.xlabel("Sepal Length (cm)")  
plt.ylabel("Frequency")  
plt.show()
```

```python
# Box plot = shows the median, quartiles, and outliers.
sns.boxplot(x="species", y="sepal_length", data=iris_df)  
plt.title("Box Plot of Sepal Length by Species")  
plt.xlabel("Species")  
plt.ylabel("Sepal Length (cm)")  
plt.show()
```

```python
#scatter plot.It only shows the overall relationship between sepal_length and sepal_width.
plt.scatter(iris_df["sepal_length"], iris_df["sepal_width"], color="red", alpha=0.6, edgecolors="black")
#alpha=0.6 â†’ Adjusts transparency (0 = fully transparent, 1 = fully opaque).
plt.xlabel("Sepal Length (cm)")  
plt.ylabel("Sepal Width (cm)")  
plt.title("Scatter Plot of Sepal Length vs Sepal Width")  
plt.show()
```

```python
# Scatter plot to differentiate flower species.
#Used to compare how different groups behave (e.g., Do all species have the same sepal length-width relationship?).
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris_df, palette="deep") #hue Colors each species differently.

plt.title("Scatter Plot of Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
#Heatmap: shows the correlation between numerical features.
# Drop non-numeric column (species)
numeric_df = iris_df.drop(columns=["species"])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True,cmap="coolwarm", linewidths=0.5)

plt.title("Heatmap of Feature Correlations")  
plt.show()
```

```python
#Pair Plot: shows scatter plots for all numerical features.
sns.pairplot(iris_df, hue="species")  
plt.show()
```
